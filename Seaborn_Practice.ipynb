# Joint plot
1 scatter and 2 marginal distribution plot

```python
iris=sns.load_dataset('iris')
iris.head()
```

```python
plt.figure(figsize=(5,1))
sns.jointplot(data = iris, x='sepal_length',y='petal_length')#default=Scatter plot
```

```python
plt.figure(figsize=(.5,1))
sns.jointplot(data = iris, x='sepal_length',y='petal_length'
              ,kind='reg',color='green')#or kind=kde,scatter,hex
```

```python
sns.jointplot(data = iris, x='sepal_length',y='petal_length'
              ,hue='species')#or kind=kde,scatter,hex
```

```python
#formatting the entire plot
sns.jointplot(data = iris, x='sepal_length',y='petal_length',hue='species'
              ,palette='gnuplot2_r',color='black')
```

```python
#formatting the individual plots/species
sns.jointplot(data = iris, x='sepal_length',y='petal_length',hue='species'
              ,joint_kws=dict(marker='+',color='black')
              ,marginal_kws=dict(color='green'))
              #,palette='gnuplot2_r',color='black')
```

```python
#height ratio and space
sns.jointplot(data = iris, x='sepal_length',y='petal_length', height=5,ratio=3,space=1,marginal_ticks=True)
```

```python
#KDE+ROG+JOINT
l=sns.jointplot(data = iris, x='sepal_length',y='petal_length',joint_kws=dict(color='black'))#color='black')
l.plot_joint(sns.kdeplot)
l.plot_joint(sns.rugplot, height=0.1,color='black')
```

# FacetGrid Plot
helps to create individual plot for each of the variable

```python
train.head()
```

```python
print(train.Outlet_Size.unique())#Outlet_Location_Type
print(train.Outlet_Location_Type.unique())
```

```python
x=sns.FacetGrid(data=train,col='')
x.map_dataframe(sns.lineplot,x='Outlet_Year',y='Sales',ci=None
               ,hue ='Outlet_Location_Type'
               ,alpha=0.5
               ,marker='+')#or HistPlot or scatterplot
```

```python
x=sns.FacetGrid(data=train,col='Outlet_Size',row='Outlet_Location_Type',sharey=False,ylim=(0,1000))
x.map_dataframe(sns.histplot,x='Sales')
x.set_axis_labels('Sales','Count')
x.set_titles(col_templates='{col_name} Size',row_templates='{row_name} Type')
```
